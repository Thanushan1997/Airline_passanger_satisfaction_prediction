```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("airline_passenger_satisfaction.csv")
```

```python
data.head()
```

```python
data.isna().any()
```

```python
data.isna().sum()
```

```python
data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
data.head()
```

```python
train, test = train_test_split(data,test_size=0.2,random_state= 20)
```

```python
train.head()
```

```python
train.shape
```

```python
test.shape
```

```python
train.isnull().sum()
```

```python
train['arrival_delay_in_minutes'].median()
```

```python
train.fillna(0, inplace=True)
train.head()
```

```python
train.isna().any()
```

```python
train.dtypes
```

```python
cols =['inflight_wifi_service','departure_arrival_time_convenient','ease_of_online_booking','gate_location','food_and_drink','online_boarding','seat_comfort','inflight_entertainment','onboard_service','leg_room_service','baggage_handling','checkin_service','inflight_service','cleanliness']
for col in cols:
    train.loc[:,col] = train.loc[:,col].astype('category')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
df = pd.read_csv("airline_passenger_satisfaction.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df.dtypes
```

```python
data = df.drop('Unnamed: 0', axis=1)
```

```python
# split into train test sets
train, test = train_test_split(data,test_size=0.2,random_state= 20)
```

```python
train.shape
```

```python
train.head()
```

```python
train.isnull().sum()
```

```python
train['arrival_delay_in_minutes'].fillna(0, inplace = True) 
```

```python
train.isnull().sum()
```

```python
cols =['inflight_wifi_service','departure_arrival_time_convenient','ease_of_online_booking','gate_location','food_and_drink','online_boarding','seat_comfort','inflight_entertainment','onboard_service','leg_room_service','baggage_handling','checkin_service','inflight_service','cleanliness']
for col in cols:
    train.loc[:,col] = train.loc[:,col].astype('object')
```
